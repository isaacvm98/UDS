## Stakeholder Summary

This memo is written for a Director of Workforce Policy at a state Department of Labor. The intended stakeholder is assumed to be comfortable interpreting applied research and labor market indicators, but not interested in econometric detail. The main priority is actionable evidence that can inform workforce planning and early-career employment policy.


## Executive Summary

This memo examines whether workers in industries with higher exposure to generative AI experienced different labor market outcomes after the arrival of GenAI tools, and whether those patterns were more pronounced for younger workers. The analysis uses CPS ASEC data from 2016-2025, links those records to an industry-level AI exposure measure (`AIIE`), and compares how employment outcomes changed after the GenAI period began.

The clearest pattern is a modest worsening in unemployment outcomes after 2023 for workers in higher-AI-exposure industries. That shift appears more pronounced for workers ages 25-34 than for older workers. By contrast, no comparable wage effect is visible in the data over this short post period.

These results should be interpreted with caution. The unemployment findings are more informative than the wage findings, but neither result should be treated as definitive proof of causality. Even so, the evidence points to a practical concern for this stakeholder: labor market conditions may be softening somewhat more for younger workers in AI-exposed sectors, which makes those workers an appropriate focus for monitoring and policy attention.


## Decisions To Be Made

Based on these findings, three near-term decisions appear especially relevant:

1. Whether to prioritize monitoring and support for younger workers in AI-exposed sectors, especially workers in the 25-34 range.
2. Whether to expand labor-market monitoring so that hiring, job-finding, and layoffs can be separated more clearly in future work.
3. Whether to treat wage changes as a secondary indicator and focus policy attention first on employment outcomes.


## The Problem

Much of the public discussion around generative AI focuses on whether AI will eliminate jobs or lower wages. For a labor market policymaker, the more practical question is narrower: are workers in more AI-exposed industries already showing signs of weaker labor market outcomes, and are those patterns more pronounced for younger workers?

That framing matters because the analysis here covers workers ages 25-34, not only first-time labor-market entrants. The relevant policy concern is therefore broader than a “first job versus layoff” distinction. The evidence is better suited to identifying whether employment outcomes are softening more for earlier-career workers in AI-exposed industries than for older workers.


## Analytical Approach

To study this question, CPS ASEC data from 2016-2025 were merged with an industry-level AI exposure measure, `AIIE`. The analysis then compares how unemployment and wage outcomes changed after the GenAI period began in industries with higher AI exposure versus industries with lower AI exposure.

The memo focuses on two outcomes:

- Unemployment
- Wage income

The analysis also looks separately at younger workers ages 25-34 to assess whether labor market changes appear more concentrated earlier in workers' careers.

Examples of relatively high-AI-exposure industries in this sample include legal services, accounting and payroll services, computer systems design, and management consulting. Lower-exposure examples include landscaping, services to buildings and dwellings, warehousing and storage, and truck transportation.

More emphasis is placed on the unemployment results because they are more consistent with the assumptions needed for this type of comparison. The wage results are still useful descriptively, but they are less persuasive as evidence of a distinct AI-related shift.

## Main Findings

### 1. Higher AI exposure is associated with a relative increase in unemployment after 2023

The main unemployment comparison suggests that unemployment outcomes worsened slightly more after 2023 in higher-AI-exposure industries than in lower-exposure industries.

This is a modest effect, not a dramatic one. It also does not mean exposed industries became high-unemployment industries overall. In the raw data, those industries still tend to have lower unemployment than less-exposed industries. The key point is that their advantage narrowed in the post period.


### 2. The unemployment shift is larger for workers ages 25-34

The age comparison suggests that the post-period unemployment shift is larger for workers ages 25-34 than for workers ages 35-64.

That pattern matters because it suggests the labor market impact is not evenly spread across age groups. At the same time, the analysis does not identify the exact mechanism. The safest conclusion is that younger workers in higher-AI-exposure industries experienced relatively weaker unemployment outcomes after 2023.


### 3. The youngest subgroup shows the largest point estimate, but the age split is not conclusive

When the younger group is split further, the estimated effect is larger for ages 25-29 than for ages 30-34. However, that difference is not statistically precise enough to support a strong claim that the youngest subgroup is distinctly more affected.

The age-split evidence is therefore best treated as suggestive. It points toward somewhat greater vulnerability among the youngest workers in the sample, but it does not establish a sharp divide within the 25-34 group.


### 4. No meaningful wage effect is detected

The wage analysis does not show evidence that wages changed differently after 2024 in higher-AI-exposure industries.

This should not be read as proof that AI has no wage effects. The post period is still short, and wage adjustments may take longer to appear. The wage analysis is also the weaker comparison in this memo because its pre-period patterns are less stable.


## Why the Unemployment Result Is More Informative Than the Wage Result

The memo includes a diagnostic check to see whether high- and low-exposure industries were already moving differently before the GenAI period began. That check is more reassuring for unemployment than for wages.

For unemployment, the pre-period differences are smaller and less systematic, although they are not absent. For wages, the pre-period differences are more persistent, which makes it harder to interpret the later wage comparison as evidence of a distinct post-GenAI shift.

## Robustness

The younger-worker unemployment pattern remains visible after dropping 2020, the most unusual COVID year. That reduces concern that the main result is being driven entirely by pandemic volatility.

At the same time, the broader caution remains the same: these checks improve confidence that the pattern is not a one-year anomaly, but they do not turn the analysis into definitive causal proof.


## Limitations

This analysis has a few important limitations.

First, the CPS is a repeated cross-section, not a panel, so the same worker cannot be followed over time. The results therefore describe changes in group-level outcomes rather than individual trajectories.

Second, `AIIE` measures potential AI exposure based on industry task content, not actual AI adoption by a specific employer.

Third, the analysis cannot cleanly distinguish between reduced hiring, weaker job-finding, and increased layoffs. It shows differential employment outcomes, but not the exact reason for them.

Fourth, the post period is still short, especially for wages.

Finally, the supporting diagnostic checks are more favorable for unemployment than for wages, but they are not strong enough to remove all concern about pre-existing differences across industries.


## Conclusion

The main takeaway is that labor market conditions appear to have weakened somewhat more after 2023 for younger workers in higher-AI-exposure industries. The unemployment comparison points in that direction, and the age breakdown suggests the pattern is stronger for workers ages 25-34 than for older workers.

That said, the memo should not be read as definitive proof that generative AI caused these changes. The evidence is better viewed as an early warning signal. For this stakeholder, the most practical implication is that younger workers in AI-exposed sectors warrant closer monitoring, while wage effects are not yet clearly visible in the data.

### Appendix A. Data Assembly Code

The code below shows how the CPS data were merged with industry-level AI exposure (`AIIE`) and occupation-level AI exposure (`AIOE`). Because CPS ASEC uses three different Census industry vintages across 2016–2025 (2012 codes for 2016–2019, 2017 codes for 2020–2024, 2022 codes for 2025), the assembly builds a single unified AIIE lookup that covers all three vintages via alias rows so every year resolves against the same table.


```python
import os
from pathlib import Path
import sys
import pandas as pd
import warnings

warnings.filterwarnings("ignore")
os.environ["PYTHONWARNINGS"] = "ignore"

for candidate in [Path.cwd(), Path.cwd().parent]:
    if (candidate / "notebook_helpers.py").exists():
        if str(candidate) not in sys.path:
            sys.path.insert(0, str(candidate))
        REPO_ROOT = candidate
        break
else:
    raise FileNotFoundError("Could not find notebook_helpers.py")

from notebook_helpers import (
    ind_to_aiie,
    occ_to_aioe,
    parse_naics_patterns,
    parse_soc_pattern,
    to_canonical_ind,
)

aiie = pd.read_excel("../data/AIOE_DataAppendix.xlsx", sheet_name="Appendix B")

crosswalk_clean = (
    pd.read_excel(
        "../data/2022-Census-Industry-Code-List-with-Crosswalk.xlsx",
        sheet_name="2022 Census Ind Code List ",
    )
    .dropna(how="all")
    .rename(
        columns={
            "Industry 2022 Description": "industry_desc",
            "2022 Census Industry Code": "census_ind_raw",
            "2022 NAICS Code": "naics_raw",
        }
    )
)

is_leaf = ~crosswalk_clean["census_ind_raw"].astype(str).str.contains("-", na=False)
crosswalk_clean = crosswalk_clean.loc[is_leaf].copy()
crosswalk_clean["IND"] = pd.to_numeric(
    crosswalk_clean["census_ind_raw"], errors="coerce"
).astype("Int64")
crosswalk_clean = crosswalk_clean.dropna(subset=["IND"])

crosswalk_clean["naics_patterns"] = crosswalk_clean["naics_raw"].apply(
    parse_naics_patterns
)

# 2017 -> 2022: ffill both sides so split- and merge-continuation rows aren't dropped,
# then dedupe (blank separator rows between groups otherwise leak duplicates).
cc_17 = pd.read_excel(
    "../data/2022-Census-Industry-Code-List-with-Crosswalk.xlsx",
    sheet_name="2022 Ind Code Changes",
    header=2,
)
cc_17[["2017 Census Industry Code", "2017 Census Industry Description"]] = cc_17[
    ["2017 Census Industry Code", "2017 Census Industry Description"]
].ffill()
cc_17[["2022 Census Industry Code", "2022 Census Industry Description"]] = cc_17[
    ["2022 Census Industry Code", "2022 Census Industry Description"]
].ffill()
cc_17 = cc_17.dropna(
    subset=["2017 Census Industry Code", "2022 Census Industry Code"]
).astype({"2017 Census Industry Code": int, "2022 Census Industry Code": int})
cc_17 = cc_17.drop_duplicates(
    subset=["2017 Census Industry Code", "2022 Census Industry Code"]
)

n_children_17 = cc_17.groupby("2017 Census Industry Code").size()
singletons_17 = n_children_17[n_children_17 == 1].index
splits_17 = n_children_17[n_children_17 > 1].index
ind_2017_to_2022 = dict(
    zip(
        cc_17.loc[
            cc_17["2017 Census Industry Code"].isin(singletons_17),
            "2017 Census Industry Code",
        ],
        cc_17.loc[
            cc_17["2017 Census Industry Code"].isin(singletons_17),
            "2022 Census Industry Code",
        ],
    )
)
ind_2017_splits = {
    parent: grp["2022 Census Industry Code"].tolist()
    for parent, grp in cc_17[
        cc_17["2017 Census Industry Code"].isin(splits_17)
    ].groupby("2017 Census Industry Code")
}

# 2012 -> 2017: CPS ASEC 2016-2019 data uses 2012 Census codes. This was an error
cc_12 = pd.read_excel(
    "../data/2017-industry-code-list-with-crosswalk.xlsx",
    sheet_name="Summary of Changes",
    header=2,
).rename(columns=lambda c: c.strip())
cc_12[["2012 Census Industry Code", "2012 Description"]] = cc_12[
    ["2012 Census Industry Code", "2012 Description"]
].ffill()
cc_12[["2017 Census Industry Code", "2017 Description"]] = cc_12[
    ["2017 Census Industry Code", "2017 Description"]
].ffill()
cc_12 = cc_12.dropna(
    subset=["2012 Census Industry Code", "2017 Census Industry Code"]
).astype({"2012 Census Industry Code": int, "2017 Census Industry Code": int})
cc_12 = cc_12.drop_duplicates(
    subset=["2012 Census Industry Code", "2017 Census Industry Code"]
)

n_children_12 = cc_12.groupby("2012 Census Industry Code").size()
singletons_12 = n_children_12[n_children_12 == 1].index
splits_12 = n_children_12[n_children_12 > 1].index
ind_2012_to_2017 = dict(
    zip(
        cc_12.loc[
            cc_12["2012 Census Industry Code"].isin(singletons_12),
            "2012 Census Industry Code",
        ],
        cc_12.loc[
            cc_12["2012 Census Industry Code"].isin(singletons_12),
            "2017 Census Industry Code",
        ],
    )
)
ind_2012_splits = {
    parent: grp["2017 Census Industry Code"].tolist()
    for parent, grp in cc_12[
        cc_12["2012 Census Industry Code"].isin(splits_12)
    ].groupby("2012 Census Industry Code")
}
```

```python
aiie_lookup = dict(zip(aiie["NAICS"].astype(str), aiie["AIIE"]))
aiie_keys_longest_first = sorted(aiie_lookup, key=len, reverse=True)

crosswalk_clean["AIIE"] = crosswalk_clean["naics_patterns"].apply(
    lambda patterns: ind_to_aiie(patterns, aiie_lookup, aiie_keys_longest_first)
)
ind_aiie = (
    crosswalk_clean.dropna(subset=["AIIE"])
    .astype({"IND": "int64"})[["IND", "industry_desc", "AIIE"]]
    .reset_index(drop=True)
)

# Add alias rows so 2017- and 2012-vintage CPS codes resolve against the same table.
# 1:1 renames/merges inherit the child's AIIE; split parents get the mean of children.
aiie_by_2022 = dict(zip(ind_aiie["IND"], ind_aiie["AIIE"]))

alias_2017 = []
for c17, c22 in ind_2017_to_2022.items():
    if c17 in aiie_by_2022:
        continue
    if c22 in aiie_by_2022:
        alias_2017.append(
            {"IND": c17, "industry_desc": "(2017 alias)", "AIIE": aiie_by_2022[c22]}
        )
for parent, kids in ind_2017_splits.items():
    vals = [aiie_by_2022[k] for k in kids if k in aiie_by_2022]
    if vals:
        alias_2017.append(
            {
                "IND": parent,
                "industry_desc": f"(2017 parent of {len(kids)} 2022 codes)",
                "AIIE": sum(vals) / len(vals),
            }
        )

aiie_by_2017 = dict(aiie_by_2022)
for row in alias_2017:
    aiie_by_2017[row["IND"]] = row["AIIE"]

alias_2012 = []
for c12, c17 in ind_2012_to_2017.items():
    if c12 in aiie_by_2017:
        continue
    if c17 in aiie_by_2017:
        alias_2012.append(
            {"IND": c12, "industry_desc": "(2012 alias)", "AIIE": aiie_by_2017[c17]}
        )
for parent, kids in ind_2012_splits.items():
    vals = [aiie_by_2017[k] for k in kids if k in aiie_by_2017]
    if vals:
        alias_2012.append(
            {
                "IND": parent,
                "industry_desc": f"(2012 parent of {len(kids)} 2017 codes)",
                "AIIE": sum(vals) / len(vals),
            }
        )

ind_aiie = pd.concat(
    [ind_aiie, pd.DataFrame(alias_2017), pd.DataFrame(alias_2012)],
    ignore_index=True,
)

df = pd.read_csv("../data/cps_00001.csv")

df["IND_2022"] = df["IND"].map(
    lambda x: to_canonical_ind(x, ind_2012_to_2017, ind_2017_to_2022)
)
df_aiie = df.merge(ind_aiie, on="IND", how="left")
```

```python
aioe = pd.read_excel("../data/AIOE_DataAppendix.xlsx", sheet_name="Appendix A")

occ_crosswalk = (
    pd.read_excel(
        "../data/2018-occupation-code-list-and-crosswalk.xlsx",
        sheet_name="2018 Census Occ Code List",
        header=0,
    )
    .dropna(how="all")
    .rename(
        columns={
            "2018 Census Title ": "occ_desc",
            "2018 Census Code": "census_occ_raw",
            "2018 SOC Code": "soc_raw",
        }
    )
)

is_leaf = ~occ_crosswalk["census_occ_raw"].astype(str).str.contains("-", na=False)
occ_crosswalk = occ_crosswalk.loc[is_leaf].copy()
occ_crosswalk["OCC"] = pd.to_numeric(
    occ_crosswalk["census_occ_raw"], errors="coerce"
).astype("Int64")
occ_crosswalk = occ_crosswalk.dropna(subset=["OCC"])

occ_crosswalk["soc_pattern"] = occ_crosswalk["soc_raw"].apply(parse_soc_pattern)

aioe_lookup = dict(zip(aioe["SOC Code"].astype(str), aioe["AIOE"]))
aioe_keys_longest_first = sorted(aioe_lookup, key=len, reverse=True)

occ_crosswalk["AIOE"] = occ_crosswalk["soc_pattern"].apply(
    lambda pattern: occ_to_aioe(pattern, aioe_lookup, aioe_keys_longest_first)
)
occ_aioe = (
    occ_crosswalk.dropna(subset=["AIOE"])
    .astype({"OCC": "int64"})[["OCC", "occ_desc", "AIOE"]]
    .reset_index(drop=True)
)
```

```python
occ_changes = (
    pd.read_excel(
        "../data/2018-occupation-code-list-and-crosswalk.xlsx",
        sheet_name="Occ Code Changes",
        header=2,
    )
    .rename(
        columns={
            "2010 Census OCC Code": "occ_2010",
            "2010 Census Title": "title_2010",
            "New 2018 Census OCC Code": "occ_2018",
        }
    )
    .dropna(how="all")
)

occ_changes[["occ_2010", "title_2010"]] = occ_changes[
    ["occ_2010", "title_2010"]
].ffill()
occ_changes = occ_changes.dropna(subset=["occ_2010", "occ_2018"]).astype(
    {"occ_2010": int, "occ_2018": int}
)

occ_2018_to_aioe = dict(zip(occ_aioe["OCC"], occ_aioe["AIOE"]))
occ_changes["AIOE"] = occ_changes["occ_2018"].map(occ_2018_to_aioe)

pre2018 = (
    occ_changes[~occ_changes["occ_2010"].isin(occ_aioe["OCC"])]
    .dropna(subset=["AIOE"])
    .groupby("occ_2010", as_index=False)
    .agg(occ_desc=("title_2010", "first"), AIOE=("AIOE", "mean"))
    .rename(columns={"occ_2010": "OCC"})
)

occ_aioe_full = pd.concat([occ_aioe, pre2018], ignore_index=True)
df_aiie = df_aiie.merge(occ_aioe_full, on="OCC", how="left")

df_aiie.to_csv("../data/data_with_aiie.csv", index=False)
```

### Appendix C. Core Estimation Code

This section includes the code used for the descriptive plots, event-study diagnostics, main difference-in-differences regressions, and the age-heterogeneity specifications.


#### Difference-in-Differences Specifications

The baseline unemployment difference-in-differences model is:

$$
\begin{aligned}
unemployed_{it} &= \beta_0 + \beta_1 AIIE_j + \beta_2 post^{unemp}_t \\
&\quad + \beta_3 \left(AIIE_j \times post^{unemp}_t\right) + \beta_4 AGE_i \\
&\quad + \sum_e \delta_e 1\{EDUC_i = e\} + \varepsilon_{it}
\end{aligned}
$$

The unemployment heterogeneity model adds a triple interaction:

$$
\begin{aligned}
unemployed_{it} &= \alpha_0 + \alpha_1 AIIE_j + \alpha_2 post^{unemp}_t + \alpha_3 early\_career_i \\
&\quad + \alpha_4 \left(AIIE_j \times post^{unemp}_t\right) \\
&\quad + \alpha_5 \left(AIIE_j \times early\_career_i\right) \\
&\quad + \alpha_6 \left(post^{unemp}_t \times early\_career_i\right) \\
&\quad + \alpha_7 \left(AIIE_j \times post^{unemp}_t \times early\_career_i\right) \\
&\quad + \alpha_8 AGE_i + \sum_e \rho_e 1\{EDUC_i = e\} + u_{it}
\end{aligned}
$$

The baseline wage difference-in-differences model is:

$$
\begin{aligned}
\log(wage_{it}) &= \gamma_0 + \gamma_1 AIIE_j + \gamma_2 post^{wage}_t \\
&\quad + \gamma_3 \left(AIIE_j \times post^{wage}_t\right) + \gamma_4 AGE_i \\
&\quad + \sum_e \kappa_e 1\{EDUC_i = e\} + \nu_{it}
\end{aligned}
$$

The wage heterogeneity model parallels the unemployment triple-interaction setup:

$$
\begin{aligned}
\log(wage_{it}) &= \eta_0 + \eta_1 AIIE_j + \eta_2 post^{wage}_t + \eta_3 early\_career_i \\
&\quad + \eta_4 \left(AIIE_j \times post^{wage}_t\right) \\
&\quad + \eta_5 \left(AIIE_j \times early\_career_i\right) \\
&\quad + \eta_6 \left(post^{wage}_t \times early\_career_i\right) \\
&\quad + \eta_7 \left(AIIE_j \times post^{wage}_t \times early\_career_i\right) \\
&\quad + \eta_8 AGE_i + \sum_e \phi_e 1\{EDUC_i = e\} + \omega_{it}
\end{aligned}
$$

Variable definitions:

- $unemployed_{it}$: indicator equal to 1 if individual $i$ is unemployed in year $t$
- $\log(wage_{it})$: log wage income for individual $i$ in year $t$
- $AIIE_j$: AI Industry Exposure for industry $j$
- $post^{unemp}_t$: indicator equal to 1 in 2023-2025
- $post^{wage}_t$: indicator equal to 1 in 2024-2025
- $early\_career_i$: indicator equal to 1 for ages 25-34
- $AGE_i$: age in years
- $1\{EDUC_i = e\}$: education-category controls

All models are estimated using CPS ASEC weights and cluster-robust standard errors at the 2022 Census industry level.

Now, we will dive into a check for parallel trends.

#### Why Check Parallel Trends?

The descriptive figures are useful for showing the broad pattern in the data, but they are not enough on their own to justify a difference-in-differences interpretation. The key identifying assumption behind the DiD design is that, absent the post-2022 expansion of generative AI, outcomes in higher- and lower-exposure industries would have continued to move along similar underlying paths.

This is why we now dive into an event-study version of the model. The event-study check asks whether the relationship between `AIIE` and the outcome was already shifting before the post period began. If those pre-period differences are small, that makes the DiD setup more credible. If they are large or systematic, then the post-period DiD estimates are harder to interpret causally because some of the gap may reflect pre-existing divergence rather than a post-2022 change.


#### Parallel Trends and Event-Study Diagnostic

Parallel trends are evaluated with an event-study specification that replaces the single post-period interaction with year-specific interactions between `AIIE` and survey year:

$$
\begin{aligned}
Y_{it} &= \theta_0 + \sum_{\tau \neq 2022} \theta_{\tau} \left(AIIE_j \times 1\{YEAR_t = \tau\}\right) \\
&\quad + \gamma_j + \sum_{\tau \neq 2022} \lambda_{\tau} 1\{YEAR_t = \tau\} \\
&\quad + \sum_k \psi_k X_{kit} + \varepsilon_{it}
\end{aligned}
$$

where $Y_{it}$ is either unemployment or log wage income, and $X_{kit}$ includes age and education controls. This is not a staggered-adoption design. The omitted year is 2022, so each coefficient shows how the relationship between `AIIE` and the outcome differs relative to 2022.

Interpretation of the diagnostic:

- If pre-period coefficients are close to zero, that supports the parallel-trends assumption.
- If pre-period coefficients are systematically different from zero, that suggests pre-existing differential trends.
- The joint pre-trend test evaluates whether all pre-period `year × AIIE` terms are zero together.

Neither outcome fully passes this diagnostic. The unemployment specification still looks more informative than the wage specification because its pre-period deviations are smaller and less systematic, but it should still be treated as suggestive rather than definitive.
